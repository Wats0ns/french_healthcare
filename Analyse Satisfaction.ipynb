```python
import pandas as pd
import plotly
import plotly.plotly as py
#import  plotly.offline  as py
import plotly.graph_objs as go
import os
%matplotlib inline
#py.offline.init_notebook_mode(connected=True)
plotly.tools.set_credentials_file(username=os.environ['PLOTLY_USERNAME'], api_key=os.environ['PLOTLY_API_KEY'])
```

```python
ds = pd.read_csv('etat_sante_general.csv')
ds = ds.set_index('état')
ds.head()
```

Une colonne sur deux représente la taille de l'échantillon, et la deuxième représente le %. On ne garde donc que la 2e pour ce graphique

```python
df = ds.copy()
for ix, column in enumerate(ds):
    if 'Unnamed' in column:
        name = ds.columns[ix - 1]
        df.drop(ds.columns[ix - 1], 1, inplace=True)
        df = df.rename(index=str, columns={column: name})
df.head()
```

```python
df.index
ds_sb = pd.DataFrame({'Très bon': df.loc['Très bon'], 'Bon': df.loc['Bon'], 'Moyen': df.loc['Moyen'], 'Mauvais': df.loc['Mauvais'], 'Très mauvais': df.loc['Très mauvais'], })
ds_sb.index
ds_sb = ds_sb.sort_index()
# Les données ne sont pas complètes avant 2007
ds_sb = ds_sb.loc['2007':]
```

```python
trace1 = dict(
    x=ds_sb.index,
    y=ds_sb['Très mauvais'],
    fill='tozeroy',
    line=dict(width=0.5,
              color='#ee1c25'),
    stackgroup='one',
    mode='lines',
    name='Très mauvaise'

)

trace2 = dict(
    x=ds_sb.index,
    y=ds_sb['Mauvais'],
    fill='tonexty',
    mode='lines',
    stackgroup='one',
    line=dict(width=0.5,
              color='#ed403c'),
    name='Mauvaise'
)

trace3 = dict(
    x=ds_sb.index,
    y=ds_sb['Moyen'],
    fill='tonexty',
    mode='lines',
    stackgroup='one',
    line=dict(width=0.5,
              color='#f58225'),
    name='Moyenne'
)


trace4 = dict(
    x=ds_sb.index,
    y=ds_sb['Bon'],
    fill='tonexty',
    mode='lines',
    stackgroup='one',
    line=dict(width=0.5,
              color='#6fbe43'),
    name='Bonne'
)

trace5 = dict(
    x=ds_sb.index,
    y=ds_sb['Très bon'],
    fill='tonexty',
    mode='lines',
    stackgroup='one',
    line=dict(width=0.5,
              color='#03a45e'),
    name='Très bonne',
    opacity=0
)
layout = go.Layout(
    showlegend=True,
    xaxis=dict(
        type='category',
        showgrid=False
    ),
    yaxis=dict(
        type='linear',
        range=[1, 100],
        dtick=20,
        ticksuffix='%',
        showgrid=False,
    ),
    title="% de Français décrivant leur santé comme:",
)
data = [trace1, trace2, trace3, trace4, trace5]
fig = dict(data=data, layout=layout)
py.iplot(fig, filename='sante_francais', validate=False)
```
